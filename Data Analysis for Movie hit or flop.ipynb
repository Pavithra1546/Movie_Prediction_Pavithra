```python
import pandas as pd
```

```python
dataset=pd.read_csv("Hit or Flop movie Dataset.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
df=dataset[dataset["Success/Failure"]==1]
df
```

# 3)What is relationship between pre-release song views and profitable movies(success)?

```python
#correlation
df.corr()
```

```python
data=df['Pre-released song views.1'].corr(df['Profit.1'])
data
```

```python
#Correlation-Relationship between 2 columns
#Pre-released song views,Profit
  #Positive correlation,26% directly proportional
```

# 4)What is relationship between Trailer views and profitable movies(success)?

```python
data=df['YouTube Views.1'].corr(df['Profit.1'])
data
```

```python
#Correlation-Relationship between 2 columns
#Trailer views,Profitable movies
#Positive correlation,21% directly proportional
```
