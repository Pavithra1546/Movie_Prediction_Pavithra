```python
#variance inflation factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    #claculating VIF
    vif=pd.DataFrame()
    vif["variable"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
    return (vif)
```

```python
calc_vif(dataset[["Investment.1","First day Box office collection.1"]])
```

```python
from scipy.stats import ttest_ind
#drop null values
dataset=dataset.dropna()
failure=dataset[dataset["Success/Failure"]==0]["First day Box office collection.1"]
success=dataset[dataset["Success/Failure"]==1]["First day Box office collection.1"]
#print(male)
ttest_ind(failure,success)
```
