```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as matlab
```

```python
dataset=pd.read_csv("Hit or Flop movie Dataset.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['Investment.1',"Profit.1", 'IMDb Rating',
       'YouTube Views.1', 'Common Background music', 'Theme background Music',
       'No.of Part Released', 'Comedy Scene',"Fast mode","Slow mode","Medium mode",
                     "Pre-released song views.1","YouTube Views.3",
       'First day Box office collection_34 Crore',
       'First day Box office collection_4.32 Crore',
       'First day Box office collection_40 Crore',
       'First day Box office collection_5 Crore',
       'First day Box office collection_5.9 Crore',
       'First day Box office collection_6 Crore',
       'First day Box office collection_60 lakh',
       'First day Box office collection_8 Crore',
       'First day Box office collection_82 Crore',
       'First day Box office collection_`10 Crore']]
dependent=dataset[['Success/Failure']]
```

```python
independent.shape
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2']}
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
#fitting the model for grid search
grid.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,y_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
